```python

@interact(a1=(1, 10, 1), b1=(1, 10, 1), c1=(1, 10, 1), a2=(1, 10, 1), b2=(1, 10, 1), c2=(1, 10, 1), cmap1=plt.colormaps(), cmap2=plt.colormaps(), space=(0, 2, 0.01))
def test_plot(a1=8, b1=8, c1=4, a2=8, b2=3, c2=4, use_data2=True, cmap1='cool', cmap2='viridis', grid=True, space=0.1):


    data1 = np.random.rand(a1, b1, c1)
    data2 = np.random.rand(a2, b2, c2)
    
    group_labels = [f'Grup-{i}' for i in range(data1.shape[0])]
    external_labels = [f'Ex-Bar-{i}' for i in range(data1.shape[1])]
    internal_labels = [f'In-Bar-{i}' for i in range(data2.shape[1])]

    

    if use_data2:
        models(data1=data1, 
               data2=data2, 
               group_labels=group_labels, 
               external_labels=external_labels, 
               internal_labels=internal_labels, 
               cmap1=[cmap1], 
               cmap2=[cmap2], 
               grid=grid, 
               theta1=10, 
               theta2=10,
               limits=[0.4, 0.6],
               rlabel_position = 0,
               space=space,
              )
    else:
        models(data1=data1, 
               # data2=data2, 
               group_labels=group_labels, 
               external_labels=external_labels, 
               cmap1=[cmap1], 
               grid=grid, 
               theta1=10, 
               limits=[0, 0.4],
               rlabel_position = 0,
               space=space,
              )
```
